```python
'''1.	Haga un programa en python que conociendo la figura geométrica que se tiene 
(Triángulo, Rectángulo o Círculo) calcule su área. Además, se tiene la siguiente información: 
'''
```

```python
#Definiendo funcion para calcular figuras 
#Recuerda las formulas para calcular areas 
#Trangulo -> se calcula su área con la formula (Base x H)/2
#Rectangulo -> Se calcula su área con la formula Base x Altura
#Circulo -> se calcula su area con la formula π r² donde el r es el radio del circulo
def figuras():
        print("Por favor elija una figura para calcular su area:") #Menu para seleccionar figura geometrica
        print("Figuras a calcular:")
        print("Para triangulo por favor digite 1.")
        print("Para Rectangulo por favor digite 2.")
        print("Para circulo por favor digite 3.")
        
        respuesta = int(input("Ingrese una opción: "))
        if respuesta <= 0:
            print("Por favor elegir entre las 3 opciones validas.")
            return
        elif respuesta >= 4 :
            print("Por favor elegir una opcion entre las 3 validas")
            return
        else:
            if respuesta == 1:# si la respuesta es 1 se calcula el área del triangulo
                print("Opción 1: Triangulo")#Se imprime la opcion elegida por el usuario
                base = float(input("Ingrese la base de la figura: "))# cargando valor de la base a la variable base
                if base <= 0:
                    print("Por favor insertar una base mayor a cero ")#Controlando que la base no sea menor o igual a cero
                    return
                else:  
                    altura = float(input("Ingrese la altura de la figura: "))
                    if altura <= 0: #verificando que la altura sea un valor valido, es decir mayor a cero
                        print("Por favor insertar un valor mayor que cero")
                        return
                    else:
                        
                        triangulo = base * altura 
                        print(f"El área calculada para los valores insertados es: {triangulo}")  
                
            elif respuesta == 2:
                print("Opción 2: Rectangulo") #Se imprime la opcion elegida por el usuario
                ancho = float(input("Ingrese el ancho: "))
                if ancho <= 0:
                    print("Por favor inserte un valor mayor que cero")
                    return
                else:
                    alto = float(input("Ingrese el valor de la altura: "))
                    if alto <=0:#verificando que la altura no sea menor que cero
                        print("Por favor ingrese un valor mayor que cero")
                        return
                    else:
                        rectangulo = alto * ancho
                print(f"El area calculada para el rectangulo con los valores insertados es: {rectangulo}")
            
            elif respuesta == 3:
                print("Opción 3: Circulo")#indicando al usuario cual opción eligio
                radio = float(input("Ingrese el radio: "))
                pi = 3.1416
                if radio <= 0:#verificando que el valor no sea menor que cero
                    print("Por favor ingresar un valor mayor que cero")
                    return
                else:
                    circulo = pi * (radio ** 2)
                print(f"El area calculada para el circulo con los valores insertados es: {circulo}")
figuras()
```
